```python
# Activating the chrome browser
driver=webdriver.Chrome("E:\Data science\FLIPROBO\web driver\chromedriver.exe") 
time.sleep(3)
```

```python
#opening the cars24 website
url = 'https://bit.ly/3HbOtY7'
driver.get(url)
```

```python
driver.find_element_by_xpath("//button[@class='btn _3BmFU AgO6q']").click()
```

```python
#Creating empty lists
brand=[]
year=[]
fuel=[]
transmission=[]
km=[]
price=[]
prod_URL=[]
```

```python
#Getting the URL of the cars
for i in range(0,500):
    url=driver.find_elements_by_xpath("//a[@class='_9Ue0B']")
    for i in url:
        prod_URL.append(i.get_attribute('href'))
    try:
        next_btn = driver.find_element_by_xpath("//button[@class='btn _3BmFU AgO6q']").click()
    except NoSuchElementException:
        pass
```

```python
prod_URL
```

```python
len(prod_URL)
```

```python
for i in prod_URL:
    driver.get(i) 
    
#Extracting car name
    try:
        brand.append(driver.find_element_by_xpath("//h2[@class='_3FpCg']").text)
    except NoSuchElementException as e:
        brand.append("-")
        
#Extracting the transmission
    try:
        transmission.append(driver.find_element_by_xpath("//p[@class='cvakB']/span").text)  
    except NoSuchElementException as e:
        transmission.append("-")
        
#Extracting the fuel consumed
    try:
        fuel.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[3]").text)
    except NoSuchElementException as e:
        fuel.append("-")
        
#Extracting km driven
    try:
        km.append(driver.find_element_by_xpath("//ul[@class='bVR0c']/li[1]").text)
    except NoSuchElementException as e:
        km.append("-") 
        

        #Extracting the price details
    try:
        price.append(driver.find_element_by_xpath("//div[@class='_7udZZ']").text.replace('',''))
    except NoSuchElementException as e:
        price.append("-")
```

```python
year = []
brand1 = []
for i in brand:
    year.append(i.split(' ')[0])
year
```

```python
car=pd.DataFrame()
car['Brand'] = brand
car['Transmission'] = transmission
car['Fuel'] = fuel
car['Kms'] = km
car['Price'] = price
car['year'] = year
car
```

### we have got a 4363 rows and 5 columns for our dataset

```python
#Saving in a csv file
car.to_csv('cars24data.csv')
```
